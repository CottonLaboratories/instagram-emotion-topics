```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)

import re
import emoji
```

```python
df = pd.read_csv('combined_dataset.csv', dtype={'text': str, 'full_name': str, 'user_name': str, 'code': str}, encoding='utf8')
```

```python
df.sample(n=20)
```

```python
#Funktion, die Links und Newlines entfernt
def cleanText(text):
    #text = re.sub('#', '', text)
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text, flags=re.MULTILINE)    
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\n \n', '', text)

    return text
```

```python
df['text'] = df['text'].astype(str)
df['text'] = df['text'].apply(cleanText)
```

```python
def extract_hashtags(text):
    hashtags = re.findall(r'#\w+', text)
    return hashtags if hashtags else None
```

```python
df['hashtags'] = df['text'].apply(extract_hashtags)
```

```python
#Funktion, die Hashtags z√§hlt und in einer Liste anlegt
def count_hashtags(hashtags_list):
    if hashtags_list is not None:
        return pd.Series(hashtags_list)
    else:
        return pd.Series([])
```

```python
#Die 20 h√§ufigsten Hashtags pro Partei werden gez√§hlt
party_hashtags = df.groupby('partei')['hashtags'].apply(count_hashtags).reset_index(name='hashtags')
party_hashtags = party_hashtags.explode('hashtags').dropna()
top_hashtags_per_party = party_hashtags.groupby('partei')['hashtags'].value_counts().groupby(level=0).head(20)
```

```python
#Top 20 werden in einem Dict angelegt und angezeigt
hashtag_dict = top_hashtags_per_party.to_dict()
hashtag_dict
```

```python
def count_hashtags(text):
    hashtags = re.findall(r'#\w+', text)
    return len(hashtags)
```

```python
df = df[df['partei'] != 'unabh√§ngig']
```

```python
#Funktion, die Emojis z√§hlt
def count_emojis(text):
    emojis = [c for c in text if emoji.is_emoji(c)]  # Verwenden von emoji.is_emoji, um zu pr√ºfen, ob ein Zeichen ein Emoji ist
    return len(emojis)
```

```python
df['hashtag_count'] = df['text'].apply(count_hashtags)
```

```python
average_hashtags_per_party = df.groupby('partei')['hashtag_count'].mean()
average_hashtags_per_party
```

```python
df['emoji_count'] = df['text'].apply(lambda x: count_emojis(x))
```

```python
average_emojis_per_party = df.groupby('partei')['emoji_count'].mean()
```
