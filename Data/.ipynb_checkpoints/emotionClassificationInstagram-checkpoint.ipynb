```python
import pandas as pd
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
import re
import torch
import ast

classifier = pipeline("text-classification", model="bert-base-german-cased-dair-ai-emotion-de-finetuned")
```

```python
df = pd.read_csv("combined_dataset_labeled.csv")
```

```python
df = df.rename(columns={"label": "topic"})
```

```python
df = df.drop(["class_probability", "view_count"], axis=1)
```

```python
df['text'] = df['text'].astype(str)
```

```python
df_10 = df.head(10)
```

```python
sequence_to_classify = df["text"]

emotion_labels = []

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# Iteriere über die Sequenz unter Verwendung von tqdm, um den Fortschritt anzuzeigen
for text in tqdm(sequence_to_classify, desc="Processing"):
    try:
        emotion_label = classifier(text)
        emotion_labels.append(emotion_label)
    except Exception as e:
        print(f"Fehler beim Verarbeiten des Textes: {e}")
        # Füge einen Platzhalterwert hinzu, um anzuzeigen, dass die Verarbeitung fehlgeschlagen ist
        emotion_labels.append(None)
```

```python
emotion_labels
```

```python
dummy_data = [{'label': 'Error', 'score': 0}]
```

```python
list_of_dicts = [item[0] if item else dummy_data[0] for item in emotion_labels]
```

```python
len(list_of_dicts)
```

```python
df
```

```python
df["class_probability"] = list_of_dicts
```

```python
df.to_csv("combined_dataset_topic_emotion_labeled.csv", index=False)
```
